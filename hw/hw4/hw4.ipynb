```python
np.random.seed(0)
X1 = np.random.choice(np.arange(1,7), size = 10000) 
X2 = np.random.choice(np.arange(1,7), size = 10000, p = [(1/16), (1/16), (3/16), (3/16), (4/16), (4/16)]) 
```

```python
Y = X1 * X2
```

```python
sns.distplot(Y, hist = True, kde = True)
plt.axvline(np.mean(Y), color = 'green')
plt.title('Distribution of the Product of r.v. X1 and X2')
plt.xlabel('product of X1 and X2')
plt.ylabel('fraction')
plt.savefig('ds100_hw4_distplot1.png');
```

```python
sns.boxplot(y = Y, showmeans = True)
plt.title('Boxplot Distribution of the Product of r.v. X1 and X2')
plt.ylabel('product of X1 and X2')
plt.savefig('ds100_hw4_boxplot1.png');
```

```python
print('Mean of sample Y:', np.mean(Y))
print('Variance of sample Y:', np.var(Y))
```
